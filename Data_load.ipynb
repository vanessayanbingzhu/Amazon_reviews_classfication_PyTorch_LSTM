```python
import pandas as pd
import numpy as np
```

```python
file_path = '/content/drive/MyDrive/Colab Notebooks/Amazon Reviews LSTM'
test_file = f'{file_path}/test.csv'
train_file = f'{file_path}/loaded_train.csv'
```

```python
dataset_test = pd.read_csv(test_file)
dataset_train = pd.read_csv(train_file)
```

```python
dataset_train.drop(columns=["Unnamed: 0"], inplace=True)
df = pd.concat([dataset_test, dataset_train], axis=0)
```

```python
sub_df = df.sample(20000)
sub_df['title_text'] = sub_df['review_title'] + ". " + sub_df['review_text']
sub_df.drop(columns=['review_title', 'review_text'], inplace=True)
sub_df = sub_df[['title_text','class_index']]
sub_df.head(2)
```

```python
sub_df.to_csv(f"{file_path}/amazon_reviews.csv", index=False)
```
